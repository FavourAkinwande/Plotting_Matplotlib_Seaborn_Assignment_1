```python

# Plot the histogram

plt.hist(df['co2'], color='blue', bins = 40, edgecolor= 'black', alpha = 0.5, label='bin width=10')

# Add a title to the plot
plt.title('Carbon Dioxide Distribution')

# Add a label to the x-axis
plt.xlabel('CO2 Emissions')

# Add a label to the y-axis
plt.ylabel("Frequency")

# Display the Histogram
plt.show()
```

Description of the CarbonDioxide Distribution Histogram:


This histogram shows the distribution of CO2 emissions across vehicles. The x-axis represents the range of CO2 emissions, which starts at 0 (less polluting vehicle) and increases towards 700 ( most polluting vehicle). The y-axis represents the frequency, or the number of vehicles that emit a specific amount of CO2 within a certain range on the x-axis. The histogram also shows a bell-shaped distribution, suggesting the data is approximately normally distributed with a peak around the 300 - 400 range with a frequency of over 350 vehicles . The distribution appears symmetric, with the frequencies decreasing evenly as we move away from the center. There are no visible outliers.


Interpretation of the CarbonDiaoxide Distribution Histogram:

```python

# Plot the histogram
plt.hist(df['ghgScore'], color='yellow', bins = 10, edgecolor= 'black', label='bin width= 30')

# Add a title to the plot
plt.title('Ghg Score Distribution')

# Add a label to the x-axis
plt.xlabel('ghgScores')

# Add a label to the y-axis
plt.ylabel("Frequency")

# Display the Histogram
plt.show()
```

Description of the Ghg Score Distribution Histogram:

The histogram represents the distribution of GHG scores of CO2 emission. The x-axis shows the GHG score ranges, and the y-axis represents the frequency of entities that fall within each range of emissions.

Showing a bell-shaped distribution, indicating that the majority of entities have moderate GHG scores. The peak of the distribution is around 4 to 6 range, suggesting that most entities in the dataset emit moderate amounts of greenhouse gases. This also implies that GHG emissions are symmetrically distributed around a central value, with fewer entities emitting very low or very high amounts of GHGs. The score tapers off gradually on both sides, meaning that extreme values are less common. There are no significant outliers, as the majority of the score is clustered around the center without extreme values on either side.

```python

# Plot the histogram
plt.hist(df['feScore'], color='green', bins = 10, edgecolor= 'black', alpha = 0.8, label='bin width= 50')

# Add a title to the plot
plt.title('Fuel-Efficiency Distribution')

# Add a label to the x-axis
plt.xlabel('feScores')

# Add a label to the y-axis
plt.ylabel("Frequency")

# Display the Histogram
plt.show()
```


Description of the Fuel-efficiency Score Distribution Histogram:

The histogram shows a bell-shaped distribution of Fuel Efficiency (FE) scores, with values ranging from 0 to 10 where most vehicles score around the middle of the FE range, typically between 4 and 6.
The peak of the histogram is around an FE score of 5, with the frequency around 1000 vehicles, indicating that a large portion of the dataset has a moderate FE score. The distribution is symmetrical, meaning the frequency of vehicles with very low and very high FE scores is relatively balanced, decreasing evenly on either side of the central peak. As we move towards the higher and lower ends of the FE score range (near 0 or 10), the frequency gradually declines, with relatively fewer vehicles scoring at the extremes. There are no significant outliers in the dataset.

```python
# Selected columns with numerical values
selected_columns = ['id', 'year', 'cylinders', 'displ', 'pv2', 'pv4', 'city', 'UCity', 'highway', 'UHighway', 'comb', 'co2', 'feScore', 'ghgScore']

# Creating a subset DataFrame with selected columns
df1 = df[selected_columns]

# Calculating the correlation matrix
corr_df1 = df1.corr()

#Plotting the heatmap using seaborn
plt.figure(figsize= (8,6))
sns.heatmap(corr_df1, cmap='coolwarm', fmt=".2f", annot=True, annot_kws={'fontsize':9})
plt.title('Heatmap Showing the Correlation of Fuel-efficiency Distribution')
plt.show()
```
